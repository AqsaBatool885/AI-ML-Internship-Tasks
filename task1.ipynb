```python
# Cell 1: Loading the Data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('iris')
print("Dataset successfully loaded!")
```

```python
#  Statistical Inspection Check for missing values and data types
print("--- Data Info ---")
df.info()

# Get the math (Mean, Max, Min)
print("\n--- Summary Statistics ---")
print(df.describe())
```

```python
# 1. Print Shape (Rows, Columns)
print(f"Shape: {df.shape}")

# 2. Print Column Names
print(f"Columns: {df.columns.tolist()}")

# 3. View first 5 rows
print(df.head())

# 4. Summary Statistics (The Math)
print(df.describe())
```

```python
# Scatter Plot & Histograms
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title("Relationship between Sepal Length and Width")
plt.show()
```

```python
df.hist(edgecolor='black', linewidth=1.2, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Box Plots (Finding Outliers)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='petal_length')
plt.title("Checking for Outliers in Petal Length")
plt.show()
```
